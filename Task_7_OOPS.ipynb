### Problem 1: Smart Light Controller
#### Problem Statement:
- Create a class to control a smart light that can be turned ON or OFF and display its status.

```python
class LightController:
    def __init__(self, action, name):
        self.action = action
        self.name = name
    def display(self, action, name):
        if self.action == "ON":
            print(self.name,"is ON")
        else:
            print(self.name,"is OFF")
light_name = input("Enter light name")
action = input("Enter action")
status = LightController(action, light_name)
status.display(action, light_name)

# class keyword is used to define a class
# __init__ is the constructor for initializing class members
# self keyword is used to access members of class
# To create object of a class in main function we use ClassName(parameter1, parameter2, ...)
```

### Problem 2: Employee ID Card System
#### Problem Statement:
- Create a class that stores employee name, ID, and department and displays ID card details.

```python
class Employee:
    def __init__(self, name, ID, dept):
        self.name = name
        self.ID = ID
        self.dept = dept
    def display(self, name, ID, dept):
        print("Employee ID Card", end="\n")
        print("Name:",self.name)
        print("ID:",self.ID)
        print("Department:",self.dept)
name = input("Employee Name: ")
ID = input("Employee ID: ")
dept = input("Department: ")
details = Employee(name, ID, dept)
details.display(name, ID, dept)
```

### Problem 3: Mobile Contact Record
#### Problem Statement:
- Create a class to store contact name and phone number and display contact information.

```python
class MobileRecord:
    def __init__(self, name, number):
        self.name = name
        self.number = number
    def display(self, name, number):
        print("Contact Saved", end="\n")
        print("Name:", self.name)
        print("Phone:", self.number)
name = input("Contact Name: ")
phone = int(input("Phone Number: "))
details = MobileRecord(name, phone)
details.display(name, phone)
```

### Problem 4: Product Price Tag Generator
#### Problem Statement:
- Create a class that stores product name and price and prints a formatted price tag.

```python
class Price:
    def __init__(self, name, amt):
        self.name = name
        self.amt = amt
    def display(self, name, amt):
        print("Product:", self.name)
        print("Price: ₹", self.amt)
name = input("Product Name: ")
amt = int(input("Price: "))
details = Price(name, amt)
details.display(name, amt)
```

### Problem 5: Movie Rating Display System
#### Problem Statement:
- Create a class that stores movie name and rating and displays movie details.

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating
    def display(self, name, rating):
        print("Movie:", self.name)
        print("Rating:",self.rating,"/5")
name = input("Movie Name: ")
rating = float(input("Rating: "))
details = Movie(name, rating)
details.display(name, rating)
```
