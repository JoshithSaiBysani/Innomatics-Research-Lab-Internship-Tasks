### Problem 1: Unique Words in a Sentence
#### Problem Statement
- Given a sentence, identify all unique words and display:
- The count of unique words
- The unique words themselves


```python
sentence = input("Enter sentence")
my_set = set(sentence)
print("Unique words count:",len(set(sentence.split()))) 
print("Unique words:",set(sentence.split()))

# split() function is used to get words from sentence
# set() function is used to convert the words into a set so duplicates are removed
# len() function is used to print the unique words count
```

### Problem 2: Highest Salary from Employee Data
#### Problem Statement:
- Store employee names and their salaries, and determine which employee has the highest salary.

```python
employees = {}
n = int(input("Enter the number of items in the dictionary"))
for i in range(n):
    name = input("Enter the employee name")
    salary = int(input("Enter employee salary"))
    employees[name] = salary
print("Employee data:", employees)
max_ans = max(employees.values()) 
for key, value in employees.items():
    if value == max_ans:
        ans = key
        break
print("Highest Salary:", ans,"-",max_ans)

# values() function is used to get the values of the dictionary
# items() is used to access the key-value pairs in the dictionary
```

### Problem 4: Count Products Above a Price Threshold
#### Problem Statement
- Count how many products have a price greater than 1000.

```python
prices = list(map(int, input("Enter the prices of products").split()))
count = 0
for price in prices:
    if price > 1000:
        count += 1
print("Products above 1000:", count)
```

### Problem 5: Calculate Attendance Percentage
#### Problem Statement
- Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).

```python
attendance = list(map(str, input("Enter the attendance").split()))
count = 0
for i in attendance:
    if i == 'P':
        count += 1
print("Attendance Percentage:", (count/len(attendance)) * 100)
```

### Problem 7: Count Character Frequency
#### Problem Statement
- Count how many times each character appears in a given string.

```python
my_str = input("Enter a string")
counts = {}
for i in my_str:
    if i in counts:
        counts[i] += 1
    else:
        counts[i] = 1
print(counts)
```

### Problem 9: Check if a Key Exists in a Dictionary
#### Problem Statement
- Check whether a specific key exists in a dictionary.

```python
employees = {}
n = int(input("Enter the number of items in the dictionary"))
for i in range(n):
    name = input("Enter the employee name")
    salary = int(input("Enter employee salary"))
    employees[name] = salary
print("Employee data:", employees)
search_key = input("Enter the employee to search")
for i in employees.keys():
    if search_key == i:
        print("Employee exists")
        break
    else:
        print("Employee does not exist")
```
