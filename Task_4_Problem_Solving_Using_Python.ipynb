### Problem 1: Mobile Recharge Validation System
#### Problem Statement
- Create a function to validate a mobile recharge request.
#### Rules:
- Recharge amount must be ≥ ₹50
- Valid plans: 199, 299, 399, 599
- User should be allowed to retry if invalid

```python
def validate_recharge():
    while True:
        amount = int(input("Enter recharge amount"))
        if amount < 50:
            print("Recharge amount should be >= ₹50. Please retry")
        if amount == 199 or amount == 299 or amount == 399 or amount == 599:
            print("Recharge successful")
            break
validate_recharge()
```

### Problem 2: Inventory Reorder Alert System
#### Problem Statement
- Given a dictionary of products and stock quantities:
- Print Reorder Alert if stock < 15
- Print Stock OK otherwise

```python
def validate_stock(products):
    for product, qty in products.items():
        if qty < 15:
            print("Reorder Alert for",product)
        else:
            print("Stock OK for",product)
        print()
products = {}
n = int(input("Enter the number of items in the dictionary"))
for i in range(n):
    item = input("Enter product name")
    quantity = int(input("Enter it's quantity"))
    products[item] = quantity
validate_stock(products)
```

### Problem 3: Student Result Processing System
#### Problem Statement
- Create a function that:
- Accepts a list of student marks
- Calculates average marks
- Prints Pass if average ≥ 50 else Fail

```python
def calculate_average(marks):
    total = 0
    for i in marks:
        total += i
    average_marks = total / len(marks)
    if average_marks >= 50:
        return "Pass"
    else:
        return "Fail"
marks = list(map(int, input("Enter marks").split()))
result = calculate_average(marks)
print(result)
```

### Problem 5: Employee Attendance Eligibility Checker
#### Problem Statement
- Create a function that:
- Accepts attendance list ("P", "A")
- Calculates attendance percentage
- Returns Eligible if ≥ 75% else Not Eligible

```python
def calculate_attendance(attendance):
    present_count = 0
    for i in attendance:
        if i == 'P':
            present_count += 1
    att_percent = (present_count / len(attendance)) * 100
    if att_percent >= 75:
        return "Eligible"
    else:
        return "Not Eligible"
attendance = list(input("Enter attendance (P/A):").upper())
result = calculate_attendance(attendance)
print(result)
```

### Problem 6: Password Strength Checker
#### Problem Statement
- Create a function to check password strength.
#### Rules:
- Minimum length = 8
- Must contain at least one digit
- Must contain at least one special character (@#$)

```python
def validate_password(password):
    if len(password) < 8:
        print("Weak password. Length must be >= 8 characters")
    for i in password:
        if i.isdigit():
            digit = True
        if i in "@#$":
            special = True
    if digit and special:
        print("Strong password")
password = input("Enter the password")
validate_password(password)
```
