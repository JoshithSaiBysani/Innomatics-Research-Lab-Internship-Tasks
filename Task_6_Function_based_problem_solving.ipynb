### Problem Statement 1: Smart Parking Lot Management System
- Design a function to manage a smart parking lot.
#### The system should:
- Accept vehicle entry and exit logs
- Calculate total parked vehicles
- Identify peak parking usage
- Alert if parking exceeds capacity

```python
def parking_lot_management(capacity, logs):
    in_count = 0
    for i in logs:
        if i == "IN":
            in_count += 1
    out_count = len(logs) - in_count
    currently_parked = in_count - out_count
    print("Currently Parked Vehicles:", currently_parked)
    if currently_parked < capacity:
        print("Parking Status: Available")
    else:
        print("Parking Status: Unavailable")
capacity = int(input("Enter parking capacity"))
logs = list(map(str, input("Enter vehicle logs").upper().split()))
parking_lot_management(capacity, logs)
```

### Problem Statement 2: Online Food Delivery Time Estimator
#### Create a function that estimates delivery time based on:
- Distance (km)
- Weather condition
- Traffic level
- Apply delays dynamically and display final ETA.

```python
def delivery_time_calculator(dist, weather, traffic):
    base_speed = 30
    base_time = (dist / base_speed) * 60
    if traffic.lower() == "high":
        base_time *= 1.4
    else:
        base_time *= 1.2
    if weather.lower() == "rainy":
        base_time *= 1.25
    else:
        base_time *= 1.1
    return int(base_time)
dist = int(input("Distance (km):"))
traffic = input("Traffic Level:")
weather = input("Weather:")
print("Estimated Delivery Time:",delivery_time_calculator(dist, weather, traffic),"minutes")
```

### Problem Statement 3: Movie Theatre Seat Occupancy Analyzer
- Build a function that analyzes seat booking data and:
- Calculates occupancy percentage
- Determines if show is Housefull
- Suggests opening additional shows

```python
def analyze_occupancy(bookings, total_seats):
    occupancy = int((len(bookings) / total_seats) * 100)
    print("Occupancy:", occupancy,"%")
    if occupancy < 25:
        print("Show Status: Many seats available")
    elif occupancy >= 25 and occupancy < 50:
        print("Show Status: Moderate occupancy")
    elif occupancy >= 50 and occupancy < 100:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Full")
total_seats = int(input("Enter total seats"))
bookings = list(map(int, input("Enter booked seats").split()))
analyze_occupancy(bookings, total_seats)
```

### Problem Statement 4: Cloud Server Load Classification System
- Create a function to classify server load based on CPU usage readings.
#### Rules:
- Average CPU < 50% → Normal
- 50%–80% → Warning
80% → Critical

```python
def load_classifier(readings):
    avg = int(sum(readings) / len(readings))
    print("Average CPU Load:", avg, "%")
    if avg < 50:
        print("Server Status: Normal")
    elif avg >= 50 and avg < 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")
readings = list(map(int, input("Enter readings").split()))
load_classifier(readings)
```

### Problem Statement 5: Smart Classroom Resource Usage Monitor
- Design a function that tracks usage of classroom resources (projector, AC, lights) and identifies overuse patterns.

```python
def monitor_usage(usage):
    threshold = {"projector": 3,
                 "ac": 6,
                 "lights": 2
                }
    overused = []
    for item, hour in usage.items():
        if hour > threshold[item]:
            overused.append(item)
    print("Overused Resources:", overused)
    if overused:
       print("Energy Alert: Yes")
    else:
        print("Energy Alert: No")
n = int(input("Enter number of items"))
usage = {}
for i in range(n):
    device = input("Enter device").lower()
    hours = int(input("Enter hours of usage"))
    usage[device] = hours
monitor_usage(usage)
```
